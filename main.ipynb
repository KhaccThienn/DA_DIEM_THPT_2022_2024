**_1. Load Data From Dataset_**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_columns", None)
```

```python
df_22 = pd.read_csv('./dataset/diem_thi_thpt_2022.csv')
df_23 = pd.read_csv('./dataset/diem_thi_thpt_2023.csv')
df_24 = pd.read_csv('./dataset/diem_thi_thpt_2024.csv')
```

**_2. Data Cleaning_**


**2.1 Data Type, NULL Values, Dupicated Value**


```python
df_22.sort_values(by='sbd', ascending=True)
```

```python
df_23.sort_values(by='sbd', ascending=True)
```

```python
df_24.sort_values(by='sbd', ascending=True)
```

_Transform Table_


```python
df_22['ma_ngoai_ngu'] = None
df_22['year'] = '2022'
df_23['year'] = '2023'
df_24['year'] = '2024'
```

```python
df_24
```

_Concat All Table_


```python
df_data_all = (
    pd.concat([df_24, df_23, df_22], ignore_index=True)
)
```

**_3. Analyze_**


**3.1 Overview**


Thống kê số lượng sinh viên theo từng năm và Tỷ lệ tăng trưởng số sinh viên theo từng năm.


```python
# Đếm số lượng sinh viên theo từng năm
student_counts_per_year = df_data_all['year'].value_counts().sort_values()

print(student_counts_per_year)

# Tính tỷ lệ tăng trưởng
growth_rate = student_counts_per_year.pct_change() * 100

# Tạo DataFrame mới với các cột cần thiết
student_growth_df = pd.DataFrame({
    "So luong thi sinh": student_counts_per_year,
    "Ty le tang truong %": growth_rate
}).reset_index().rename(columns={"index": "year"})

# Chuyển đổi cột year sang kiểu số
student_growth_df['year'] = pd.to_numeric(student_growth_df['year'])
```

```python
# Biểu đồ 1: Số lượng thí sinh qua các năm
fig1, ax1 = plt.subplots(figsize=(12, 6))
ax1.bar(student_growth_df['year'], student_growth_df['So luong thi sinh'], color='cornflowerblue')
ax1.set_xlabel('Năm')
ax1.set_ylabel('Số Lượng Thí Sinh (triệu người)')
ax1.set_title('Số Lượng Thí Sinh Qua Các Năm')
# Thêm nhãn hiển thị số lượng sinh viên lên từng cột với định dạng số thập phân
for container in ax1.containers:
    ax1.bar_label(container, label_type='edge', fmt='%.0f')
ax1.grid(axis='y', linestyle='--', linewidth=0.6)

# Hiển thị biểu đồ 1
plt.show()

# Biểu đồ 2: Tỷ lệ tăng trưởng qua các năm
fig2, ax2 = plt.subplots(figsize=(12, 6))
ax2.bar(student_growth_df['year'], student_growth_df['Ty le tang truong %'], color='orange')
ax2.set_xlabel('Năm')
ax2.set_ylabel('Tỷ Lệ Tăng Trưởng (%)')
ax2.set_title('Tỷ Lệ Tăng Trưởng Qua Các Năm')
for container in ax2.containers:
    ax2.bar_label(container, label_type='edge', fmt='%.0f')
ax2.grid(axis='y', linestyle='--', linewidth=0.6)

# Hiển thị biểu đồ 2
plt.show()
```

Biểu đồ biểu diễn sự biến thiên điểm trung bình qua từng năm


```python
import numpy as np
from scipy.interpolate import make_interp_spline

data = df_data_all
# Chuyển đổi cột year từ object sang numeric
data['year'] = pd.to_numeric(data['year'])

# Khai báo mảng các năm
x_year = [2022, 2023, 2024]

# Định nghĩa các trường môn học và tiêu đề tương ứng
subjects = ['toan', 'ngu_van', 'ngoai_ngu', 'vat_li', "hoa_hoc", 'sinh_hoc', 'lich_su', 'dia_li', 'gdcd']
titles = ['Toán', 'Văn', 'Ngoại Ngữ', 'Vật Lý', 'Hóa Học', 'Sinh Học', 'Lịch Sử', 'Địa Lý', 'GDCD']

# Tính toán điểm trung bình của từng môn qua các năm
mean_score_by_year = data.groupby('year')[subjects].mean().reset_index()

# Kiểm tra xem có giá trị NaN hoặc vô cực trong dữ liệu không
print(mean_score_by_year.isnull().sum())

# Thay thế các giá trị NaN bằng giá trị trung bình của cột tương ứng hoặc giá trị khác phù hợp
mean_score_by_year = mean_score_by_year.fillna(mean_score_by_year.mean())

# Kiểm tra lại sau khi xử lý NaN
print(mean_score_by_year.isnull().sum())

# Vẽ biểu đồ đường cho từng môn
plt.figure(figsize=(12, 6))

for subject in subjects:
    x = mean_score_by_year['year']
    y = mean_score_by_year[subject]
    
    # Tạo các điểm có tọa độ x mới để làm smoother line
    x_new = np.linspace(x.min(), x.max(), 300)
    
    # Tạo spline để làm mịn
    spl = make_interp_spline(x, y, k=2)
    y_smooth = spl(x_new)
    
    plt.plot(x_new, y_smooth, label = subject, linewidth = 2.5, alpha = 1.0)
    
plt.title('Điểm trung bình các môn qua các năm')
plt.xlabel('Năm')
plt.ylabel('Điểm trung bình')
plt.xticks(x_year)
plt.legend(loc = 'upper right', bbox_to_anchor = (1.3, 1))
plt.grid(False)
plt.tight_layout()
plt.show()
```

Histogram biểu diễn phổ điểm của từng môn học theo từng năm


```python
data = df_data_all

# Chuyển đổi cột year từ object sang numeric
data['year'] = pd.to_numeric(data['year'])

selected_years = [2022, 2023, 2024]
recent_years = data[data['year'].isin(selected_years)]

# Lọc dữ liệu cho 4 năm gần đây nhất.
recent_four_years = data[data['year'].isin(data['year'].unique()[-5:])]

# Tạo histogram cho điểm Toán, Văn và Ngoại Ngữ trong 4 năm gần đây

fig, axes = plt.subplots(3, 9, figsize=(20, 8), sharex=True)

# Định nghĩa các trường môn học, tiêu đề và màu histogram tương ứng
subjects = ['toan', 'ngu_van', 'ngoai_ngu', 'vat_li', 'hoa_hoc', 'sinh_hoc', 'lich_su', 'dia_li', 'gdcd']
titles = ['Toán', 'Văn', 'Ngoại Ngữ', 'Lý', 'Hóa', 'Sinh', 'Lịch Sử', 'Địa Lý', 'GDCD']
colors = ['tomato', 'lightskyblue', 'orange', 'steelblue', 'cornflowerblue', 'olive', 'lavender', 'silver', 'cyan']

# Sắp xếp các năm gần đây theo thứ tự tăng dần
recent_four_years_sorted = sorted(recent_years['year'].unique())

# # Vẽ histogram tổng cho từng năm, và từng môn học
for i, year in enumerate(recent_four_years_sorted):
    for j, (subject, color) in enumerate(zip(subjects, colors)):
        sns.histplot(recent_years[recent_years['year'] == year][subject], bins=20, kde=True, ax=axes[i, j], edgecolor = 'black', color=color)
        axes[i, j].set_title(f'{titles[j]} {year}')
        axes[i, j].set_xlabel(f'Điểm {titles[j]}')
        axes[i, j].set_ylabel("#")

plt.tight_layout()
plt.show()
```

Chi tiết từng histogram


```python
data = df_data_all
sub = 'toan'

data['year'] = pd.to_numeric(data['year'])

recent_four_years = data[data['year'].isin(data['year'].unique()[-3:])]
score = list(range(0, 11, 1))

fig, axes = plt.subplots(3, 1, figsize=(8, 8), sharex=True)

recent_four_years_sorted = sorted(recent_four_years['year'].unique())
colors = ['steelblue', 'coral', 'seagreen']

for i, (year, color) in enumerate(zip(recent_four_years_sorted, colors)):
    sns.histplot(recent_four_years[recent_four_years['year'] == year][sub], bins=20, kde=True, ax=axes[i], edgecolor='black', color=color)
    axes[i].set_title(f"{year}")
    axes[i].set_xlabel('Diem Toan')
    axes[i].set_ylabel('So Luong')
    
    # Tinh toan cac thong tin thong ke
    year_data = recent_four_years[recent_four_years['year'] == year][sub].dropna()
    mean = year_data.mean()
    median = year_data.median()
    std_dev = year_data.std()
    
    p70 = year_data.quantile(0.70)
    p80 = year_data.quantile(0.80)
    p90 = year_data.quantile(0.90)
    p93 = year_data.quantile(0.93)
    p95 = year_data.quantile(0.95)
    p97 = year_data.quantile(0.97)
    p99 = year_data.quantile(0.99)
    
    text_str = (f'Mean: {mean:.2f} \nMedian: {median:.2f} \nSTD: {std_dev:.2f} \n'
                f'P70: {p70:.2f} \nP80: {p80:.2f} \nP90: {p90:.2f} \n'
                f'P93: {p93:.2f} \nP95: {p95:.2f} \nP97: {p97:.2f} \nP99: {p99:.2f}')
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    axes[i].text(1.05, 0.5, text_str, transform=axes[i].transAxes, fontsize=7.5, verticalalignment='center', horizontalalignment='left', bbox=props)

fig.suptitle('Phân bố điểm môn Toán qua các năm', fontsize=16)
plt.xticks(score)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Assuming df_data_all is already defined and loaded with your data
df_data_all['year'] = pd.to_numeric(df_data_all['year'])

# Lựa chọn tổ hợp cần phần tích
combo = 'A'

# Chia khung điểm từ 10 đến 30, mỗi khoảng 1 điểm
bins = np.arange(11, 31, 0.5)

# Lọc dữ liệu cho các năm 2023 và 2024
df_2023 = df_data_all[df_data_all['year'] == 2023].copy()
df_2024 = df_data_all[df_data_all['year'] == 2024].copy()

# Tạo bảng đếm số lượng học sinh trong mỗi khung điểm cho từng năm
df_2023.loc[:, 'bins'] = pd.cut(df_2023[combo], bins)
df_2024.loc[:, 'bins'] = pd.cut(df_2024[combo], bins)

count_2023 = df_2023['bins'].value_counts().sort_index()
count_2024 = df_2024['bins'].value_counts().sort_index()

# Tính toán sự biến thiên trung bình số học sinh và khoảng tin cậy
mean_diff = count_2024.mean() - count_2023.mean()
std_diff = np.sqrt(count_2024.var() + count_2023.var())
confidence_interval = stats.norm.interval(0.95, loc=mean_diff, scale=std_diff / np.sqrt(len(count_2023)))

print(f"Sự chênh lệch trung bình số học sinh: {mean_diff}")
print(f"Khoảng tin cậy 95% cho sự chênh lệch trung bình: {confidence_interval}")

plt.figure(figsize=(20, 6))
plt.bar(count_2023.index.astype(str), count_2023.values, width=0.6, label="2023", align='center', alpha=0.5, color='lightskyblue')
plt.bar(count_2024.index.astype(str), count_2024.values, width=0.6, label="2024", align='edge', alpha=0.5, color='tomato')

plt.xlabel('Khung Điểm')
plt.ylabel('Số Lượng Học Sinh')
plt.title(f'So sánh số lượng học sinh trong các khung điểm tổ hợp {combo} giữa các năm 2023 và 2024')
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```
